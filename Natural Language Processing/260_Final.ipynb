```python
import nltk
from nltk.corpus import treebank
from nltk.tag import DefaultTagger, UnigramTagger, BigramTagger, TrigramTagger

# Load and split data
sentences = treebank.tagged_sents()
test_data = sentences[:3000]
train_data = sentences[3000:]

# Define the backoff chain
t0 = DefaultTagger('NN')
t1 = UnigramTagger(train_data, backoff=t0)
t2 = BigramTagger(train_data, backoff=t1)
t3 = TrigramTagger(train_data, backoff=t2)

# Evaluation
print(f"Default Accuracy: {t0.evaluate(test_data):.4f}")
print(f"Unigram Accuracy: {t1.evaluate(test_data):.4f}")
print(f"Bigram Accuracy:  {t2.evaluate(test_data):.4f}")
print(f"Trigram Accuracy: {t3.evaluate(test_data):.4f}")
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.pipeline import Pipeline
from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

# Prepare data
X = [" ".join(s) for s in treebank.sents()]
y = [s[0][1] for s in treebank.tagged_sents()] # Using first word's tag as category

X_train, X_test = X[3000:], X[:3000]
y_train, y_test = y[3000:], y[:3000]

models = [
    ('Naive Bayes', MultinomialNB()),
    ('Logistic Regression', LogisticRegression(max_iter=1000)),
    ('SVM', SVC())
]

for name, model in models:
    pipeline = Pipeline([
        ('vectorizer', CountVectorizer()),
        ('classifier', model)
    ])
    pipeline.fit(X_train, y_train)
    predictions = pipeline.predict(X_test)
    print(f"{name} Accuracy: {accuracy_score(y_test, predictions):.4f}")
```
